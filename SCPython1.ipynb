```python
#SOLUTION
import numpy as np

def cosine(v1, v2):
    return np.dot(v1,v2)/(np.linalg.norm(v1)*np.linalg.norm(v2))

#EXAMPLE
v1 = np.array([3,1,-2,4])
v2 = np.array([-1,3,4,2])
print(cosine(v1, v2))
```

Normal distribution is very often used to model some features. It has two parameters $\mu$ and $\sigma$ and PDF is given by $f(x) = \frac{1}{\sigma \sqrt{2 \pi}}\cdot e^{-\frac{(x-\mu)^2}{2\sigma^2}}$

```python
from math import pi, e

def f(x, mu = 0, sigma = 1):
    return (1/(sigma*(2*pi)**0.5))*e**(-((x-mu)**2)/(2*sigma**2))

x = [-6+0.1*i for i in range(200)]

y1 = [f(elem) for elem in x]  #N(0,1)
y2 = [f(elem, sigma = 2) for elem in x]   #N(0,2)
y3 = [f(elem, mu=3, sigma = 2) for elem in x]  #N(3,2)

plt.plot(x,y1, c = "red")
plt.plot(x,y2, c = "blue")
plt.plot(x,y3, c = "orange")
plt.show()
```

Let's assume we observe a random variable $X$ in two groups $L_1$ and $L_2$. According to the Bayesian Theorem

<br> 
<center> $P(L_1|X) = \frac{P(X|L_1)P(L_1)}{P(X)}$ and $P(L_2|X) = \frac{P(X|L_2)P(L_2)}{P(X)}$</center>

<br>

The higher the probability $P(L_i|X)$ is, the more prone we are to consider $L_i$ as a proper group. By investigating the ratio

 <br>
<center> $\frac{P(L_1|X)}{P(L_2|X)}$ </center>

which is
<br>
<center> $\frac{P(X|L_1)P(L_1)}{P(X|L_2)P(L_2)}$ </center>
one may determine the most probable choice.


Based on the training data, we approximate the probabilities $P(X|L_i)$ and $P(L_i)$ - we need to assume the distribution function (for example, normal distribution).

<h4> Exercise10: Define a PDF for 2D normal distribution given by $f(X) = \frac{1}{2\pi\sqrt{\det(\Sigma)}}\cdot \exp(-\frac{1}{2}(X-\mu)^T\Sigma^{-1}(X-\mu))$, where $X = [X_1, X_2]$, $\mu = [\mu_1,\mu_2]$ and $\Sigma$ is a covariance matrix. Both $\mu$ and $\Sigma$ are parameters.

```python
def f(X, mu, sigma):
    t = X - mu
    a = np.transpose(t)
    a = np.dot(a, np.linalg.inv(sigma))
    a = np.dot(a, t)
    return (1/(2*pi*(np.linalg.det(sigma))**0.5))*np.exp(-0.5*a)

#example

print(f(X = np.array([1,2]), mu = np.array([1,2]), sigma = np.array([[1,0],[0,1]])))
```

<h4> Exercise11: We consider two features (represented as a vector) $[X_1,X_2]$ in a population. It appears its distribution may be characterized by normal distribution with $\mu=[1,2]$ and $$\Sigma = \begin{bmatrix} 1 & 1.1 \\ 1.1 & 4  \end{bmatrix}$$ Plot 2D PDF

```python
#SOlution

def f(X, mu, sigma):
    t = X - mu
    a = np.transpose(t)
    a = np.dot(a, np.linalg.inv(sigma))
    a = np.dot(a, t)
    return (1/(2*pi*(np.linalg.det(sigma))**0.5))*np.exp(-0.5*a)


fig = plt.figure(figsize = (12,10))
ax = plt.axes(projection='3d')

x = np.arange(-4, 6, 0.2)
y = np.arange(-3, 7, 0.2)

X, Y = np.meshgrid(x, y)
Z = np.zeros((len(x), len(y)))
for i in range(len(x)):
    for j in range(len(y)):
        coor = np.array([x[i],y[j]])     
        Z[i,j] = f(coor, mu = np.array([1,2]), sigma = np.array([[1, 1.1], [1.1, 4]]))
        

surf = ax.plot_surface(X, Y, Z, cmap = plt.cm.cividis)
ax.view_init(elev=30, azim=80)
ax.set_xlabel('x', labelpad=20)
ax.set_ylabel('y', labelpad=20)
ax.set_zlabel('z', labelpad=20)

fig.colorbar(surf, shrink=0.5, aspect=8)

plt.show()
```
